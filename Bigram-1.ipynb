```python
# making the dictionary to save the pair of token of words 
b={}
for w in words[:]:
    ch = ['<S>']+ list(w) + ['<E>']
    for ch1, ch2 in zip(ch, ch[1:]):
        bigram = (ch1, ch2) # creating touple 
        b[bigram] = b.get(bigram , 0)+ 1 # updating the count
        print(ch1, ch2)
                               
```

```python
import torch
''' NOW INSTEAD OF <S> AND <E> WE ARE GONNA USE A DOT(.)
    SO THE TOTAL NO. OF LETTERS WILL BE 27 INSTEAD OF 28'''
N = torch.zeros((27,27), dtype = torch.int32)
```

```python
# String to int encoding 

chars  = sorted(list(set(''.join(words))))
''' the abv are using <S> and <E> to show start and end but now we are using dot to rep both start and end '''
s_to_i = {s:i+1 for i,s in enumerate(chars)}  # here we are starting from 1 --> a
s_to_i['.'] = 0 # here we are gonna rep. the dot with a 0
s_to_i
```

```python
# int. to string decoding

i_to_s = {i:s for s,i in s_to_i.items()}
i_to_s
```

```python
# counting the no. of times a pair is appearing 
for w in words:
    chs = ['.']+ list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        N[ix1, ix2] += 1
```

<h2> CHECKING THE PERFORMANCE OF THE MODEL</h2>

```python
 for w in words[:3]:
    chs = ['.']+ list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        prob = P[ix1,ix2] # probabilities that the model assigns to every one of these bigrams in the datasets 
        print(f'{ch1}{ch2}: {prob:.4f}')
#'''so here we have emma , olivia and ava probabilities assigned by the bigram model'''
```

<h3> Likelihood</h3>
For a model to be good the product of the probab. should be as high as possible <br>
for convinence we use <b>log likelihood</b> cuz' log works great with the small values 

```python
log_likelihood = 0.0
n = 0

for w in words[:3]:
    chs = ['.']+ list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        prob = P[ix1,ix2] # probabilities that the model assigns to every one of these bigrams in the datasets 
        logprob = torch.log(prob) #logarithm of prob
        log_likelihood += logprob
        n += 1
        print(f'{ch1}{ch2}: {prob:.4f} {logprob: .4f}')

print(f'{log_likelihood = }')
negative_log_likelihood = -log_likelihood
print(f'{negative_log_likelihood = }')
print(f'{negative_log_likelihood/n}') # the loss fun. assigned by the model LOWER the value BETTER the quality of the model
```
